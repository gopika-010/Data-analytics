```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
print(x)
print(y)
```

```python
#LINE CHART
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line chart")
sns.lineplot(x=x,y=y,color="yellow")
plt.show()
```

```python
#SCATTR PLOT
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("scatter chart")
sns.scatterplot(x=x,y=y,marker='o',color='blue')
plt.show()
```

```python
x1=np.linspace(0,np.pi,128)
x2=np.linspace(0,np.pi,128)
y1=np.sin(x1)
y2=np.cos(x2)
```

```python
#MULTIPLE LINE PLOT
sns.lineplot(x=x1,y=y1,color="purple")
sns.lineplot(x=x2,y=y2,color="green")
```

```python
import pandas as pd
df=pd.read_csv("Employee.csv")
df
```

```python
#COLOR PALETTE
sns.color_palette("deep")
sns.color_palette("bright")
sns.color_palette("muted")
sns.color_palette("pastel")
sns.color_palette("dark")
sns.color_palette("colorblind")
```

```python
#COUNT PLOT
sns.set_theme(style="whitegrid")
sns.countplot(data=df,palette="bright")
plt.xticks(rotation=30)
plt.show()
```

```python
#HISTOGRAM
sns.set_theme(style="dark")
sns.histplot(x="JoiningYear",data=df, kde=True, color="orange")
plt.show()

```

```python
#BOX PLOT
sns.boxplot(x="JoiningYear",y="Age",data=df,palette="pastel")
plt.show()
```

```python
df=pd.read_csv("Employee.csv")
```

```python
df
```

```python
numeric_data = df.select_dtypes(include='number')
```

```python
#CLUSTER GRAPH
sns.clustermap(numeric_data, cmap="viridis", standard_scale=1)
plt.show()
```

```python
#STRIP GRAPH
sns.stripplot(x="Gender",y="Age",data=df)
plt.show()
```

```python
#VIOLIN GRAPH
sns.violinplot(x="Gender", y="Age", data=df)
plt.title("Age Distribution by Gender")
plt.show()
```

```python
#SEASONAL GRAPH
year_counts = df.groupby("JoiningYear").size().reset_index(name="Count")
sns.lineplot(x="JoiningYear", y="Count", data=year_counts, marker="o")
plt.title("Year-wise Joining Trend")
plt.show()
```
