```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,7,13)
y=x**2
print(x)
print(y)
```

```python
#HEXBIN PLOT
plt.hexbin(x, y, gridsize=10)
plt.colorbar()

```

```python
#POLAR PLOT
plt.polar(x,y)
```

```python
#ANNOTATIONS
plt.annotate('Point', xy=(x[1],y[1]))
```

```python
#GRID & AXIS
plt.grid(True)
plt.xlim(0,10)
plt.ylim(0,15)
```

```python
#SUBPLOT
fig=plt.figure(figsize=(10,6))
axes=fig.subplots(3,3)
```

```python
x1=np.linspace(0,5,13)
x2=np.linspace(1,7,13)
y1=x1**2
y2=x2**2
print(x1)
print(y1)
print(x2)
print(y2)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.plot(x1,y1,'g.')
plt.subplot(1,2,2)
plt.plot(x2,y2,'k-.')
```

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.plot(x1, y1, 'r-.')
plt.plot(x2, y2, 'b--')
plt.show()
```

```python
from mpl_toolkits import mplot3d
```

```python
#3D PLOT
fig = plt.figure(figsize= (10,8))
ax = plt.axes(projection = '3d')
```

```python
#3D LINE PLOT
fig = plt.figure(figsize= (10,8))
ax = plt.axes(projection = '3d')
xline=np.linspace(0,20,100)
yline=np.sin(xline)
zline=np.cos(xline)
ax.plot3D(xline,yline,zline,'blue')
ax.set_title("3D Line Plot")
plt.show()
```

```python
#3D SCATTER PLOT
fig = plt.figure(figsize= (10,8))
ax = plt.axes(projection = '3d')
xline=np.linspace(0,20,100)
yline=np.sin(xline)
zline=np.cos(xline)
ax.scatter(xline,yline,zline)
ax.set_title("3D Line Plot")
plt.show()
```

```python
#3D WIREFRAME PLOT
def f(x, y):
    return np.sin(x) + np.cos(y)

x = np.outer(np.linspace(1,10),np.ones(10))
y = x.copy()

X, Y = np.meshgrid(x, y)
z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, z, color='purple')
ax.set_title("3D wireframe plot")
plt.show()
```
